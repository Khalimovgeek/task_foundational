```python
import numpy as np

tf_embedding = tf.keras.layers.Embedding(input_dim = 10,
                                         output_dim = 4)

input_ids = tf.constant([[1,2,3], [4,5,0]])

output_vectors = tf_embedding(input_ids)
print(output_vectors.shape)
print(tf_embedding.get_weights()[0].shape)
```
